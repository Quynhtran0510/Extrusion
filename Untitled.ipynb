```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
from time import time
```

```python
Job= ("111222", "111222","111222", "111222","111222", "123456","123456","123456","123456","123456")
Date = ("2020-12-14 01:19:00", "2020-12-14 08:01:00", "2020-12-14 12:23:00", "2020-12-14 14:06:00", "2020-12-14 20:00:00", "2020-12-15 02:38:00", "2020-12-15 06:07:00", "2020-12-15 09:58:00", "2020-12-15 12:42:00", "2020-12-15 16:25:00")
Date = pd.to_datetime(Date)
FT = (0, 12000, 12000, 10000, 10000, 8000, 12000, 12000, 12000, 70000)
data = list(zip(Job,Date,FT))
df = pd.DataFrame(data = data, columns = ['Job','Date','FT'])
df
```

```python
df['FT2'] = ''
for i in range(int(len(df)-1)):
    if df['Date'][i].strftime("%H:%M:%S") < "07:00:00" and df['Date'][i+1].strftime('%H:%M:%S') > "07:00:00":
        df['FT2'][i] = df['FT'][i] + (df['FT'][i+1]/(df['Date'][i+1] - df['Date'][i]).total_seconds())*((pd.to_datetime(df['Date'][i].strftime('%x'))+timedelta(hours = 7)) - df['Date'][i]).total_seconds()
        df['FT2'][i+1] = (df['FT'][i+1]/(df['Date'][i+1] - df['Date'][i]).total_seconds())*((df['Date'][i+1] - (pd.to_datetime(df['Date'][i].strftime('%x'))+ timedelta(hours = 7))).total_seconds())
    elif df['Date'][i].strftime("%H:%M:%S") < "19:00:00" and df['Date'][i+1].strftime('%H:%M:%S') > "19:00:00":
        df['FT2'][i] = df['FT'][i] + (df['FT'][i+1]/(df['Date'][i+1] - df['Date'][i]).total_seconds())*((pd.to_datetime(df['Date'][i].strftime('%x'))+timedelta(hours = 19)) - df['Date'][i]).total_seconds()
        df['FT2'][i+1] = (df['FT'][i+1]/(df['Date'][i+1] - df['Date'][i]).total_seconds())*((df['Date'][i+1] - (pd.to_datetime(df['Date'][i].strftime('%x'))+ timedelta(hours = 19))).total_seconds())
    else:
        df['FT2'][i+1] = df['FT'][i+1]
```
